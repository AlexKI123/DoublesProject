```python
from scipy.special import gamma
import numpy as np
import matplotlib.pyplot as plt

# Parameters
ns = np.logspace(np.log10(1),np.log10(10000),5).astype(int)
num_trials = 1000  # Number of trials to average over
mean = 0.1
```

```python
# Parameters specific to exponential distribution
lambd = 1/mean  # Rate parameter for the exponential distribution

# Function to perform the experiment with exponential distribution
def average_max_exponential(n, trials, rate):
    max_values = []
    for _ in range(trials):
        samples = np.random.exponential(1/rate, n)
        max_values.append(np.max(samples))
    return np.mean(max_values)

# Calculate the average maximum for different n
average_max_values = [average_max_exponential(n, num_trials, lambd) for n in ns]

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(ns, average_max_values, "o")
plt.plot(ns, (0.5772156649+np.log(ns))/lambd)
plt.xscale("log")
plt.yscale("log")
plt.title('Average of the Maximum of n Exponential Random Variables')
plt.xlabel('Number of Samples (n)')
plt.ylabel('Average Maximum Value')
plt.grid(True)
plt.show()
```

```python
# Parameters specific to Pareto distribution
alpha = 2.5  # Shape parameter for the Pareto distribution
xm = (mean * (alpha - 1)) / alpha     # Scale parameter 

# Function to perform the experiment with Pareto distribution
def average_max_pareto(n, trials, alpha, xm):
    max_values = []
    for _ in range(trials):
        samples = (np.random.pareto(alpha, n) + 1) * xm
        max_values.append(np.max(samples))
    return np.mean(max_values)

# Calculate the average maximum for different n using Pareto distribution
average_max_values_pareto = [average_max_pareto(n, num_trials, alpha, xm) for n in ns]

# Plotting the results for Pareto distribution
plt.figure(figsize=(10, 6))
plt.plot(ns, average_max_values_pareto, "o")
plt.plot(ns, ns**(1/alpha)*xm*gamma((alpha-1)/alpha))
plt.xscale("log")
plt.yscale("log")
plt.title('Average of the Maximum of n Pareto Random Variables')
plt.xlabel('Number of Samples (n)')
plt.ylabel('Average Maximum Value')
plt.grid(True)
plt.show()
```

```python
# Parameters specific to exponential distribution
lambd = 1/mean  # Rate parameter for the exponential distribution

# Parameters specific to Pareto distribution
alpha = 2.5  # Shape parameter for the Pareto distribution
xm = (mean * (alpha - 1)) / alpha     # Scale parameter 

# Function to perform the experiment with exponential distribution
def average_max_exponential(n, trials, rate):
    max_values = []
    for _ in range(trials):
        samples = np.random.exponential(1/rate, n)
        max_values.append(np.max(samples))
    return np.mean(max_values)

# Function to perform the experiment with Pareto distribution
def average_max_pareto(n, trials, alpha, xm):
    max_values = []
    for _ in range(trials):
        samples = (np.random.pareto(alpha, n) + 1) * xm
        max_values.append(np.max(samples))
    return np.mean(max_values)


# Calculate the average maximum for different n
average_max_values = [average_max_exponential(n, num_trials, lambd) for n in ns]
average_max_values_pareto = [average_max_pareto(n, num_trials, alpha, xm) for n in ns]

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(ns, average_max_values, "o", color="C0", label="Exp")
plt.plot(ns, (0.5772156649+np.log(ns))/lambd, color="C0")

plt.plot(ns, average_max_values_pareto, "o", color="C1", label="Pareto")
plt.plot(ns, ns**(1/alpha)*xm*gamma((alpha-1)/alpha), color="C1")

plt.xscale("log")
plt.yscale("log")
plt.title('Average of the Maximum of n Random Variables')
plt.xlabel('Number of Samples (n)')
plt.ylabel('Average Maximum Value')
plt.grid(True)
plt.legend()
plt.show()
```

# Ratio

```python
c = 1.68
ns = np.logspace(np.log10(1),np.log10(100000),6).astype(int)
nd = np.round(ns*c,0).astype(int)
num_trials = 10000  # Number of trials to average over
```

```python
# Parameters specific to exponential distribution
lambd = 1/mean  # Rate parameter for the exponential distribution

# Function to perform the experiment with exponential distribution
def average_max_exponential(n, trials, rate):
    max_values = []
    for _ in range(trials):
        samples = np.random.exponential(1/rate, n)
        max_values.append(np.max(samples))
    return np.mean(max_values)

# Calculate the average maximum for different n
average_max_values_singles = np.array([average_max_exponential(n, num_trials, lambd) for n in ns])
average_max_values_doubles = np.array([average_max_exponential(n, num_trials, lambd) for n in nd])


# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(ns, average_max_values_doubles/average_max_values_singles, "o")
plt.plot(ns, [1+np.log(1.68)/(0.5772156649+np.log(i)) for i in ns], "-")
plt.xscale("log")
plt.yscale("log")
plt.title('Ratio of the Maximum of n Exponential Random Variables')
plt.xlabel('Number of Samples (n)')
plt.ylabel('Average Maximum Value')
plt.grid(True)
plt.show()
```

```python
ns = np.logspace(np.log10(1),np.log10(100000),6).astype(int)
nd = np.round(ns*1.68,0).astype(int)

# Parameters specific to Pareto distribution
alpha = 2.5  # Shape parameter for the Pareto distribution
xm = (mean * (alpha - 1)) / alpha     # Scale parameter 

# Function to perform the experiment with exponential distribution
def average_max_pareto(n, trials, alpha, xm):
    max_values = []
    for _ in range(trials):
        samples = (np.random.pareto(alpha, n) + 1) * xm
        max_values.append(np.max(samples))
    return np.mean(max_values)

# Calculate the average maximum for different n
average_max_values_singles = np.array([average_max_pareto(n, num_trials, alpha, xm) for n in ns])
average_max_values_doubles = np.array([average_max_pareto(n, num_trials, alpha, xm) for n in nd])


# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(ns, average_max_values_doubles/average_max_values_singles, "o")
plt.plot(ns, [c**(1/alpha) for i in ns], "o")
plt.xscale("log")
plt.yscale("log")
plt.title('Ratio of the Maximum of n Pareto Random Variables')
plt.xlabel('Number of Samples (n)')
plt.ylabel('Average Maximum Value')
plt.grid(True)
plt.show()
```
