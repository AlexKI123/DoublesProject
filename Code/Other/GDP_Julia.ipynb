# GPD

```julia
meanS = 0.1
meanD = 0.08
shape = 0.4
scaleS = meanS *(1 - shape)
scaleD = meanD *(1 - shape)
fS = rand(GeneralizedPareto(0, scaleS, shape), 1+ns) .+ 1.
fD = rand(GeneralizedPareto(0, scaleD, shape), nd) .+ 1.
f = vcat(fS, fD);
```

```julia
println(mean(fS))
println(mean(fD))
```

```julia
meanS = 0.1
shape = 0.4
scale = meanS *(1 - shape)
ns = 200
fS = rand(GeneralizedPareto(0, scale, shape), 1+ns) 
mean(fS)
```

```julia
p = histogram(fS, bins=100, normalize=:pdf, alpha=0.75, 
          xlabel="Values of f", ylabel="Frequency")
```

```julia
lambda = 1/meanS  # Define lambda for the Exponential PDF

# Define the range for plotting the PDF
x_vals = 0:0.0001:maximum(fS)  # Adjust the step and range as needed

# Calculate the Exponential PDF
exp_pdf = lambda * exp.(-lambda * x_vals)

plot!(p, x_vals, exp_pdf, line=(2, :red), label="Exponential PDF")
display(p)
```

# Maximum distribution

```julia
# Pareto
ns = 120
meanParetoS = 0.1
shape = 2.5
scaleS = (meanParetoS * (shape - 1)) / shape
num_simulations = 10000  # Number of simulations to perform

# Function to generate random samples and find their maximum
function simulate_maxima(ns, shape, scaleS, num_simulations)
    maxima = Float64[]
    for _ in 1:num_simulations
        fSP = rand(Pareto(shape, scaleS), ns) .+ 1.
        push!(maxima, maximum(fSP))
    end
    return maxima
end

# Simulate
maxima = simulate_maxima(ns, shape, scaleS, num_simulations)

# Calculate the mean of the maxima
mean_max = mean(maxima)
println("Mean of the maxima: $mean_max")

# Plot the distribution of the maxima
histogram(maxima, bins=30, label="Maxima distribution", alpha=0.7)
title!("Distribution of Maximum Values")
xlabel!("Maximum value")
ylabel!("Frequency")
```

```julia
# Shifted Pareto
ns = 120
meanParetoS = 0.1
shape = 2.5
scaleS = meanParetoS*(shape-1)
num_simulations = 100000  # Number of simulations to perform

# Function to generate random samples and find their maximum
function simulate_maxima(ns, shape, scaleS, num_simulations)
    maxima = Float64[]
    for _ in 1:num_simulations
        fSP = rand(Pareto(shape, scaleS), ns) .- scaleS .+ 1.
        push!(maxima, maximum(fSP))
    end
    return maxima
end

# Simulate
maxima = simulate_maxima(ns, shape, scaleS, num_simulations)

# Calculate the mean of the maxima
mean_max = mean(maxima)
println("Mean of the maxima: $mean_max")

# Plot the distribution of the maxima
histogram(maxima, bins=30, label="Maxima distribution", alpha=0.7)
title!("Distribution of Maximum Values")
xlabel!("Maximum value")
ylabel!("Frequency")
```
